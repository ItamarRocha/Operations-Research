# Plotting curves

```python
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

* In this section we'll plot the complexitys that are more common in our daily studies.

```python
y = np.arange(0,5,0.1)
x = np.arange(0,5,0.1)
plt.plot(x,np.ones(x.shape), label = '1') #constant
plt.plot(x,np.log(y), label = 'log(n)') # log
plt.plot(x,y, label = 'n') ## n
plt.plot(x,y * np.log(y),label ='n * log(n)')
plt.plot(x, y**2, label = 'n^2')
plt.legend();
```

* Now lets compare the bigger complexitys to have a notion. Its important to bear in mind that those function plotted bellow grow much faster than the ones we plotted in the last graph.

```python
from scipy.special import factorial
```

```python
y = np.arange(0,5,0.1)
x = np.arange(0,5,0.1)
plt.plot(x, y**3, label = 'n^3')
plt.plot(x, y**4, label = 'n^4')
plt.plot(x, 2**y, label = '2^y')
plt.plot(x, factorial(y), label = 'n!')
plt.legend();
```

* With this range we may conclude that the $n^4$ is the ones that grows faster. Although, taking into consideration a range till 10 we can see thats not the case.

```python
y = np.arange(0,10,1)
x = np.arange(0,10,1)
plt.plot(x, y**3, label = 'n^3')
plt.plot(x, y**4, label = 'n^4')
plt.plot(x, 2**y, label = '2^y')
plt.plot(x, factorial(y), label = 'n!')
plt.legend();
```

* The factorial grows much faster than the other 3. In fact, we cannot compute at all till the 200 number in factorial with our python librarys.

* Just in order to compare the other 3 in a bigger range, lets exclude the factorial.

```python
y = np.arange(0,20,1)
x = np.arange(0,20,1)
plt.plot(x, y**3, label = 'n^3')
plt.plot(x, y**4, label = 'n^4')
plt.plot(x, 2**y, label = '2^y')
plt.legend();
```

> With that we can conclude that, although $n^4$ and $n^3$ may start with some advantage, $2^y$ grows much faster.
